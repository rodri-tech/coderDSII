# Contexto general de las Operaciones Bancarias y su relación con las Compras

Las operaciones bancarias y las transacciones de compras están intrínsecamente relacionadas y forman parte esencial del sistema financiero.

**Operaciones Bancarias**

Las operaciones bancarias incluyen una amplia gama de servicios financieros proporcionados por instituciones como bancos y cooperativas de crédito. Estos servicios abarcan, entre otros:

- Cuentas Corrientes y de Ahorro: Permiten a los clientes depositar y retirar dinero, además de gestionar sus finanzas diarias.

- Préstamos y Créditos: Incluyen préstamos personales, hipotecarios y líneas de crédito, que permiten a los consumidores financiar grandes compras y proyectos.

- Tarjetas de Crédito y Débito: Facilitan las transacciones diarias y permiten a los clientes realizar compras sin necesidad de llevar efectivo.

- Transferencias y Pagos: Permiten a los clientes enviar y recibir dinero, así como pagar facturas y otros servicios.

**Relación con las Compra**
- Pagos Electrónicos:
Tarjetas de Crédito y Débito: La mayoría de las compras se realizan mediante tarjetas, que están vinculadas a las cuentas bancarias de los clientes. Las transacciones son procesadas electrónicamente y registradas en tiempo real.

- Transferencias Bancarias: Algunas compras, especialmente las de mayor valor, se realizan mediante transferencias bancarias directas.

- Comercio Electrónico:
Compras Online: El comercio electrónico ha crecido exponencialmente, y los bancos han desarrollado servicios específicos para facilitar las compras en línea, como la autenticación de dos factores para mayor seguridad.

- Programas de Recompensas y Lealtad:
Muchos bancos ofrecen programas de recompensas vinculados a las compras con tarjeta de crédito, donde los clientes ganan puntos, millas u otros beneficios que pueden canjear por productos o servicios.

- Prevención de Fraudes:
Detección de Fraude: Dado el aumento de transacciones electrónicas, la detección y prevención de fraudes se ha convertido en una prioridad. Los bancos utilizan modelos de machine learning para identificar transacciones sospechosas y proteger a los clientes.

- Seguridad y Regulación:
Regulación Financiera: Los bancos deben cumplir con regulaciones estrictas para garantizar la seguridad de las transacciones y proteger la privacidad de los clientes.

- Tecnología de Seguridad: Se implementan tecnologías avanzadas, como la encriptación y la tokenización, para proteger los datos de las transacciones.

```python
# Generamos un mapa con las ciudades de nuestro dataset y los montos de las transacciones relevadas en las mismas para mostrar cómo está distribuida nuestra muestra.

import plotly.express as px
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter

# Función para obtener las coordenadas de una ciudad
def get_coordinates(city):
    geolocator = Nominatim(user_agent="myGeocoder")
    location = geolocator.geocode(city + ', USA')
    if location:
        return location.latitude, location.longitude
    else:
        return None, None

def plot_transactions_bubble_map(df):
    # Agrupamos los datos por ciudad y sumar las transacciones
    city_data = df.groupby('City')['Transaction_Amount'].sum().reset_index()

    # Obtenemos las coordenadas de cada ciudad
    geocode = RateLimiter(get_coordinates, min_delay_seconds=1)
    city_data[['Latitude', 'Longitude']] = city_data['City'].apply(lambda city: pd.Series(geocode(city)))

    # Filtramos ciudades con coordenadas válidas
    city_data = city_data.dropna(subset=['Latitude', 'Longitude'])

    # Creamos un bubble map
    fig = px.scatter_mapbox(city_data,
                            lat='Latitude',
                            lon='Longitude',
                            size='Transaction_Amount',
                            hover_name='City',
                            hover_data={'Latitude':False, 'Longitude':False, 'Transaction_Amount':True},
                            color_continuous_scale=px.colors.cyclical.IceFire,
                            size_max=15,
                            zoom=4)

    # Establecemos el estilo del mapa
    fig.update_layout(mapbox_style="open-street-map")
    fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})

    return fig

fig = plot_transactions_bubble_map(df)
fig.show()
```

Con esta visualización de Mapa de Burbujas podemos identificar que tenemos una distribución similiar en los montos de transacciones relevadas en las distintas ciudades y a su vez visualmente podemos identificar una distribución cardinal equitativa de ciudades dentro de los Estados Unidos, es decir tenemos 3 ciudades del este, 3 del centro y 4 del oeste del país, abarcando así toda la región.

```python
# Buscaremos con este gráfico de telaraña ver si tenemos una tendencia marcada entre los métidos de pago, las ciudades y los montos de las transacciones.

def plot_average_transactions_polar(df):
    # Calculamos el promedio de transacciones por método de pago y ciudad
    mean_values = df.groupby(['Payment_Method', 'City'])['Transaction_Amount'].mean().unstack()

    # Creamos las categorías (ciudades)
    categories = mean_values.columns
    angles = np.linspace(0, 2 * np.pi, len(categories), endpoint=False).tolist()
    angles += angles[:1]
    fig, ax = plt.subplots(figsize=(10, 10), subplot_kw={'polar': True})

    for payment_method in mean_values.index:
        values = mean_values.loc[payment_method].tolist()
        values += values[:1]  # Cerramos el círculo
        ax.plot(angles, values, label=payment_method)
        ax.fill(angles, values, alpha=0.1)

    ax.set_yticks(np.arange(0, mean_values.max().max(), step=mean_values.max().max()/6))
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(categories)
    plt.title("Comparación Promedio de Transacciones por Método de Pago y Ciudad")
    plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1.05))
    plt.show()

plot_average_transactions_polar(df)
```

El gráfico de telaraña nos muestra que todas las ciudades se encuentran equilibradas en cuanto a los métodos de pagos por monto de transacciones. La superposición del gráfico en los picos que señalan las ciuidades no muestra que haya tendencias definidas entre estas variables.

```python
def plot_fraud_bar_chart(df):
    # Agrupar los datos por categoría y flag de fraude, sumando los montos de las transacciones
    grouped_data = df.groupby(['Category', 'Fraud_Flag'])['Transaction_Amount'].sum().reset_index()

    # Crear un gráfico de barras
    fig = px.bar(grouped_data,
                 x='Category',
                 y='Transaction_Amount',
                 color='Fraud_Flag',
                 title='Monto de Transacciones Fraudulentas por Categoría',
                 labels={'Category': 'Categoría', 'Transaction_Amount': 'Monto de Transacciones', 'Fraud_Flag': 'Fraude'},
                 template='plotly_white',
                 barmode='group')

    return fig

# Ejemplo de uso
# df = pd.read_csv('ruta/a/tu/dataset.csv') # Cargar tu dataset aquí
fig = plot_fraud_bar_chart(df)
fig.show()
```

Contamos con un dataset balanceado tanto en los montos de transacciones como en la distribución de nuestra variable ojetivo. Esto puede ayudar al modelo predictivo.

```python
def plot_correlation_heatmap(df):
    # Seleccionar las columnas numéricas para la correlación
    numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns

    # Determinación de la matriz de correlación
    correlation_matrix = df[numeric_columns].corr()

    # Creamos un heatmap de la matriz de correlación
    plt.figure(figsize=(12, 8))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
    plt.title('Mapa de Calor de la Correlación entre las Variables')
    plt.show()

plot_correlation_heatmap(df)
```

Con este mapa de correlación vemos que no contamos con variables que se correlaciones fuertemente, lo que puede ser un problema para desarrollar el modelo. En este caso debemos ver si esto es limitante para el desarrollo de nuestro modelo.
